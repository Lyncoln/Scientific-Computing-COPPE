```python
def getRIC(Sr,Sg,Sb,r):
    sigma_all = np.sum(Sr)+np.sum(Sg)+np.sum(Sb)
    sigma_r = np.sum(Sr[:r,:r])+np.sum(Sg[:r,:r])+np.sum(Sb[:r,:r])
    
    return(sigma_r/sigma_all)
    
```

```python
A = imread("image.png")
j = 0

for r in (5,20,140,280,560):
    #Red
    Ur, Sr, VTr, infor = sci_la.dgesdd(A[:,:,0])
    Sr = np.diag(Sr)
#     XapproxR = np.matmul(Ur[:,:r], Sr[:r,:r])
    XapproxR = sci_blas.dgemm(1,Ur[:,:r], Sr[:r,:r])
    XapproxR =sci_blas.dgemm(1,XapproxR, VTr[:r,:])
    #Green
    Ug, Sg, VTg, infog = sci_la.dgesdd(A[:,:,1])
    Sg = np.diag(Sg)
    XapproxG = sci_blas.dgemm(1,Ug[:,:r], Sg[:r,:r])
    XapproxG = sci_blas.dgemm(1,XapproxG, VTg[:r,:])
    #Blue
    Ub, Sb, VTb, infob = sci_la.dgesdd(A[:,:,2])
    Sb = np.diag(Sb)
    XapproxB = sci_blas.dgemm(1,Ub[:,:r], Sb[:r,:r])
    XapproxB = sci_blas.dgemm(1,XapproxB, VTb[:r,:])
    
    
    #Figure Plot
    plt.figure(j+1,figsize=(10,10))
    ric = round(getRIC(Sr,Sg,Sb,r),2)
    plt.title(f'r = {str(r)}, RIC = '+str(ric))
    plt.axis('off')
    rgb = np.dstack((XapproxR,XapproxG,XapproxB))
    plt.imshow((rgb * 255).astype(np.uint8))
    j += 1
    
    plt.show()



    
```

```python
import seaborn as sns
sns.set_theme(style="whitegrid")
acu = np.cumsum((np.diag(Sr)+np.diag(Sg)+np.diag(Sr))/np.sum(np.diag(Sr)+np.diag(Sg)+np.diag(Sr)))
```

```python
fig, (ax1,ax2) = plt.subplots(nrows=2, ncols=1, figsize=(10,10),sharex=True)
ax1.semilogy(np.diag(Sr)+np.diag(Sg)+np.diag(Sr))
ax1.set_ylabel('σ')
ax1.set_title('Valores singulares')
ax2.plot(acu)
ax2.set_xlabel('m')
ax2.set_ylabel('RIC')
ax2.hlines(y = 0.99, xmin = 0, xmax = 768,linestyles="--",color = "gray",alpha=0.75)
ax2.annotate("1º Maior que 0.99 -> 560",
            (np.where(acu>=0.99)[0][0],0.99),
             xytext=(np.where(acu>=0.99)[0][0]+20,0.90),
             arrowprops=dict(facecolor='black',
                             width=2,
                             headwidth=10)
            )
ax2.scatter(x = np.where(acu>=0.99)[0][0],y = 0.99,c='red')
plt.show()
```
